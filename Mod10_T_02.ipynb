````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%matplotlib widget
````

````python
pr = pd.read_csv('previsao_de_renda.csv')
pr.head()
````

````python
pp = sns.pointplot(data=pr, x='posse_de_imovel', y='renda', errorbar=('ci', 95))
plt.show()
````

````python
plt.close()
pp = sns.pointplot(data=pr, x='posse_de_veiculo', y='renda', errorbar=('ci', 95))
plt.show()
````

````python
plt.close()
pp = sns.pointplot(data=pr, x='posse_de_imovel', y='renda', hue='posse_de_veiculo', errorbar=('ci', 95))
plt.show()
````
